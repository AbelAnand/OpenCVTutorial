```python
#Hue Saturation Value
#import packages for our tutorial
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#open the original image
tutorial_image = cv2.imread('/content/drive/MyDrive/AbelFolder/spacex.jpg')
#BGR2RGB
tutorial_image_rgb = cv2.cvtColor(tutorial_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(tutorial_image_rgb)
```

```python
#converting the image to HSV
tutorial_image_hsv = cv2.cvtColor(tutorial_image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(tutorial_image_hsv)
```

```python
#spliting the hsv
h, s, v = cv2.split(tutorial_image_hsv)
```

```python
plt.figure(figsize=[20, 5])
plt.subplot(141);plt.imshow(h);plt.title("hue channels")
plt.subplot(142);plt.imshow(s);plt.title("stauration channels")
plt.subplot(143);plt.imshow(v);plt.title("value channels")
```

```python
#for saving
cv2.imwrite('spacex_img.png', tutorial_image_rgb)
```

```python
#cropping image
plt.imshow(tutorial_image_rgb)
```

```python
croping_region = tutorial_image[150:1500, 1000:1300]
```

```python
plt.imshow(croping_region[:, :, ::-1])
```

```python
#resize the image
resizing_img_2x = cv2.resize(croping_region, None, fx=2, fy=2)
plt.imshow(resizing_img_2x[:, :, ::-1])
```

```python
#method2 using height and width
height = 200
width = 200
dimension = (height, width)
image_resize = cv2.resize(croping_region, dsize=dimension, interpolation=cv2.INTER_AREA)
plt.imshow(image_resize[:, :, ::-1])
```

```python
#height and width
croping_region.shape
```

```python
#to find thew height
width = 100
aspect_ratio = width/croping_region.shape[1]
aspect_ratio
```

```python
height = int(croping_region.shape[0]*aspect_ratio)
height
```

```python
aspect_ratio = (width, height)
image_resize = cv2.resize(croping_region, dsize=aspect_ratio, interpolation=cv2.INTER_AREA)
plt.imshow(image_resize[:, :, ::-1])
```
