## Conditional Statements

```python
movie_data.original_language == "en"
```

```python
#how to use conditional statement
movie_data.loc[movie_data.original_language == "en"]
```

```python
movie_data.loc[(movie_data.original_language == "en") & (movie_data.crew == "['Christopher Nolan']")]
```

```python
movie_data.loc[(movie_data.original_language == "en") | (movie_data.crew == "['Christopher Nolan']")]
```

```python
#isin
movie_data.loc[movie_data.original_language.isin(["zh", "hi"])]
```

```python
movie_data.loc[movie_data.overview.notnull()]
```

```python
movie_data.loc[movie_data.overview.isnull()]
```
