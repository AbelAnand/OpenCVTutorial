```python
#modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
orginal_image = cv2.imread('/content/drive/MyDrive/AbelFolder/falcon9.jpg')
plt.imshow(orginal_image[:, :, ::-1])
```

```python
#starting point, endingpoint
#thickness
#color
#drawing a line using cv2 the syntax is:
"""
cv2.line(image, pt1, pt2, color, thickness, line_type)
cv2.LINE_4, cv2.LINE_8, cv2.LINE_AA
"""
```

```python
cv2.line(org_img_line_8, (900, 500), (1200, 500), (0, 255, 0), 10, cv2.LINE_8)
plt.imshow(org_img_line_8[:, :, ::-1])
```

```python
org_img_line_aa = orginal_image.copy()
cv2.line(org_img_line_aa, (900, 500), (1200, 500), (0, 255, 0), 10, cv2.LINE_AA)
plt.imshow(org_img_line_aa[:, :, ::-1])
```

```python
#circle
#center
#radius
#color
#thickness
#linetype
org_img_copy_cir = orginal_image.copy()
```

```python
cv2.circle(org_img_copy_cir, (1050, 600), 100, (255, 0, 0), 3, cv2.LINE_AA)
plt.imshow(org_img_copy_cir[:, :, ::-1])
```

```python
#rectangle
#corners - top-left vertex, bottom right vertex
#color
#thickness
#linetype
org_image_copy_rect = orginal_image.copy()
```

```python
cv2.rectangle(org_image_copy_rect, (630, 20), (820, 380), (0, 0, 255), 5, cv2.LINE_4)
plt.imshow(org_image_copy_rect[:, :, ::-1])
```

```python
#adding text on image
#text
#bottom-left vertex
#font
#fontscale
#color
#thickness
#line_type
org_img_copy_text = orginal_image.copy()
text = "Falcon 9 ->"
fontScale = 2
fontFace = cv2.FONT_HERSHEY_SIMPLEX
color = (145, 255, 200)
fontThickness = 3
```

```python
cv2.putText(org_img_copy_text, text, (180, 150), fontFace, fontScale, color, fontThickness, cv2.LINE_AA)
plt.imshow(org_img_copy_text[:, :, ::-1])
```

```python
#basic image enhancement using mathematical operations
orginal_image_enhance = cv2.imread("/content/drive/MyDrive/AbelFolder/ww2.jpg")
plt.imshow(orginal_image_enhance[:, :, ::-1])
```

```python
print(orginal_image_enhance)
```

```python
#shape of the image
print(orginal_image_enhance.shape)
```

```python
#matrix using numpy
matrix = np.ones((200, 400, 3), dtype="uint8")*50
print(matrix)
```

```python
#creating a matrix using image shape
matrix_img = np.ones((orginal_image_enhance.shape), dtype="uint8") * 50
print(matrix_img)
```

```python
plt.imshow(matrix_img[:, :, ::-1])
```

## Adding brightness ADD and SUBTRACT

```python
bgr_image_bri = cv2.add(orginal_image_enhance, matrix_img) #brighter
bgr_image_dar = cv2.subtract(orginal_image_enhance, matrix_img) #darker
plt.figure(figsize=(10, 10))
plt.subplot(131);plt.imshow(bgr_image_dar[:, :, ::-1]);plt.title("darker")
plt.subplot(132);plt.imshow(orginal_image_enhance[:, :, ::-1]);plt.title("Orginal image")
plt.subplot(133);plt.imshow(bgr_image_bri[:, :, ::-1]);plt.title("brighter")
```

## Contrast - Multiplication

```python
#numpy
orginal_image_enhance
```

```python
#typeconvertion
float_img = np.float64(orginal_image_enhance)
float_img
```

```python
#two matrices
matrix1 = np.ones(orginal_image_enhance.shape)*1.2
matrix2 = np.ones(orginal_image_enhance.shape)*0.5
plt.imshow(matrix2)
```

```python
img_cont_dark = np.uint8(cv2.multiply(float_img, matrix2))
img_cont_bright = np.uint8(cv2.multiply(float_img, matrix1))
```

```python
img_cont_dark
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(131);plt.imshow(img_cont_dark[:, :, ::-1]);plt.title("Dark contrast")
plt.subplot(132);plt.imshow(orginal_image_enhance[:, :, ::-1]);plt.title("Orginal image")
plt.subplot(133);plt.imshow(img_cont_bright[:, :, ::-1]);plt.title("Bright contrast")
```
