```python
b, g, r = cv2.split(astro_img_1)
```

```python
plt.figure(figsize=[20, 5])
plt.subplot(141);plt.imshow(b, cmap="gray");plt.title("blue channels")
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("green channels")
plt.subplot(143);plt.imshow(r, cmap="gray");plt.title("red channels")
```

```python
#merged_image b, g, r
merged_image = cv2.merge((b, g, r))
plt.imshow(merged_image)
```

```python
#merging image g, b, r
merged_image = cv2.merge((g, b, r))
plt.imshow(merged_image)
```
